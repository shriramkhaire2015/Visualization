# 1. Pie Chart

```python
cause = ['Chronic Diseases','Unintentional Injuries', 'Alzheiners','Infuenza and Pneumonia', 'Sepsis','Others']
Percentile = [22,5,4,2,1,26]
```

```python
plt.figure(figsize=(10,10))
explode = (0.05,0,0,0,0,0)
plt.pie(Percentile, labels=cause,colors=['hotpink','g', 'b', '#4CAF50','red','yellow'], explode=explode, autopct='%1.1f%%',startangle=90)
plt.axis('equal')
plt.title('Ohio State 2012 : Leading Cause of Death', fontdict={'family':'serif','color':'black','size':16})
plt.legend(title='Cause',loc='upper right',bbox_to_anchor=(1.30, 1), borderaxespad=0)
plt.show()
```

# 2. Donut Chart

```python
plt.figure(figsize=(10,10))
explode = (0.05,0.05,0.05,0.05,0.05,0.05)
# Pie Chart
plt.pie(Percentile, labels=cause,colors=['hotpink','g', 'b', '#4CAF50','red','yellow'], explode=explode, autopct='%1.1f%%',startangle=90,pctdistance=0.85)
plt.axis('equal') 

# draw circle
centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()

# Adding circle in pie chart
fig.gca().add_artist(centre_circle)

plt.title('Ohio State 2012 : Leading Cause of Death', fontdict={'family':'serif','color':'black','size':16})
plt.legend(title='Cause',loc='upper right',bbox_to_anchor=(1.30, 1), borderaxespad=0)
plt.show()
```

# 3. Bar Charts

### Simple Bar Chart

```python
x = np.array(['A','B','C','D','E']) 
y = np.array([4,9,6,10,8])
```

```python
style.use('ggplot')
ax = plt.figure(figsize=(10,6),facecolor='pink',edgecolor='#00FFFF')
plt.bar(x, y, width=0.3, color='cornflowerblue', edgecolor='red', linewidth=0.5, tick_label=x)
for i in range(len(x)):
    plt.text(i, y[i], y[i], ha='center', va='bottom')
plt.xlabel('x', fontdict={'family':'serif','color':'olive','size':15})
plt.ylabel('y', fontdict={'family':'serif','color':'olive','size':15})
plt.title('Bar Plot', fontdict={'family':'serif','color':'green','size':16})
plt.xticks(color='black')
plt.yticks(color='black')
plt.ylim((0,12))
plt.show()
```

### Horizantal Bar Chart

```python
style.use('ggplot')
ax = plt.figure(figsize=(10,6), facecolor='orange',edgecolor='#00FFFF')
hbar = plt.barh(x, y, 0.3, color='cyan', edgecolor='green', linewidth=0.5, tick_label=x)
plt.xlabel('y', fontdict={'family':'serif','color':'red','size':15})
plt.ylabel('x', fontdict={'family':'serif','color':'red','size':15})
plt.title('Horizantal Bar Plot', fontdict={'family':'serif','color':'green','size':16})
plt.xticks(color='black')
plt.yticks(color='black')
plt.bar_label(hbar, fmt='%.2f', color= 'black', padding=3)
plt.show()   
```

### Stacked Bar Chart

```python
x = np.array(['A','B','C','D','E']) 
y1 = np.array([4,9,6,10,8])
y2 = np.array([10,30,15,25,18])
y3 = np.array([22,36,28,32,26])
```

```python
style.use('seaborn-deep')
ax = plt.figure(figsize=(10,6), facecolor='cyan', edgecolor='#00FFFF')
bar1 = plt.bar(x, y1, 0.3, color='yellow')
bar2 = plt.bar(x, y2, 0.3, bottom=y1, color='hotpink')
for r1, r2 in zip(bar1, bar2):
    h1 = r1.get_height()
    h2 = r2.get_height()
    plt.text(r1.get_x() + r1.get_width() / 2., h1 / 2., '%d' % h1, ha='center', va='bottom', color='blue', fontweight='bold')
    plt.text(r2.get_x() + r2.get_width() / 2., h1 + h2 / 2., '%d' % h2, ha='center', va='bottom', color='blue', fontweight='bold')
plt.xticks(color='red')
plt.yticks(color='red')
plt.legend(['y1', 'y2'], title='Y')
plt.title('Stacked Bar Chart', fontdict={'family':'serif','color':'Green','size':16, 'fontweight':'bold'})
plt.xlabel('X', fontdict={'family':'serif','color':'orange','size':16, 'fontweight':'bold'})
plt.ylabel('Y', fontdict={'family':'serif','color':'orange','size':16, 'fontweight':'bold'})
plt.show()
```

```python
style.use('dark_background')
ax = plt.figure(figsize=(10,6), facecolor='cyan', edgecolor='#00FFFF')
bar1 = plt.bar(x, y1, 0.3, color='yellow')
bar2 = plt.bar(x, y2, 0.3, bottom=y1, color='hotpink')
bar3 = plt.bar(x, y3, 0.3, bottom=y1+y2, color='red')
for r1, r2, r3 in zip(bar1, bar2, bar3):
    h1 = r1.get_height()
    h2 = r2.get_height()
    h3 = r3.get_height()
    plt.text(r1.get_x() + r1.get_width() / 2., h1 / 2., '%d' % h1, ha='center', va='bottom', color='blue', fontweight='bold')
    plt.text(r2.get_x() + r2.get_width() / 2., h1 + h2 / 2., '%d' % h2, ha='center', va='bottom', color='blue', fontweight='bold')
    plt.text(r3.get_x() + r3.get_width() / 2., h1 + h2 + h3 / 2., '%d' % h3, ha='center', va='bottom', color='blue', fontweight='bold')
plt.xticks(color='red')
plt.yticks(color='red')
plt.legend(['y1', 'y2', 'y3'], title='Y')
plt.title('Stacked Bar Chart', fontdict={'family':'serif','color':'Green','size':16, 'fontweight':'bold'})
plt.xlabel('X', fontdict={'family':'serif','color':'orange','size':16, 'fontweight':'bold'})
plt.ylabel('Y', fontdict={'family':'serif','color':'orange','size':16, 'fontweight':'bold'})
plt.show()
```

## Multiple Bar Charts

```python
years = ['2017', '2018', '2019', '2020']
cse = [74,82,89,79]
it = [58,72,76,61]
ece = [67,74,81,68]
eee = [53,45,57,41]
```

```python
style.use('dark_background')
ax = plt.figure(figsize=(10,6), facecolor='m', edgecolor='gold')

cse_bar = np.arange(len(years))
it_bar=[i+0.2 for i in cse_bar]
ece_bar=[i+0.2 for i in it_bar]
eee_bar=[i+0.2 for i in ece_bar]

bar1 = plt.bar(cse_bar, cse, width=0.2, color='b', label='CSE')
bar2 = plt.bar(it_bar, it, width=0.2, color='orange', label='IT')
bar3 = plt.bar(ece_bar, ece, width=0.2, color='g', label='ECE')
bar4 = plt.bar(eee_bar, eee, width=0.2, color='darkred', label='EEE')
plt.title('PLACEMENT COMPARISION', fontdict={'family':'serif','color':'yellow','size':16})
plt.xlabel('Years')
plt.ylabel('Placement Counts')
plt.legend(title='Department')
plt.xticks(cse_bar+0.2+0.2/2, years)
plt.bar_label(bar1, padding=3)
plt.bar_label(bar2, padding=3)
plt.bar_label(bar3, padding=3)
plt.bar_label(bar4, padding=3)
plt.show()
```
